```python
import os, json
from dotenv import load_dotenv
from google import genai

load_dotenv()
client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])

resp = client.models.generate_content(
    model="gemini-1.5-flash",
    contents="Write one sentence about attention mechanisms.",
    config={
        "temperature": 0,
        "response_logprobs": True,
        "logprobs": 5,
    },
)

print(resp.text)
print(json.dumps(resp.candidates[0].logprobs_result, indent=2))
```
